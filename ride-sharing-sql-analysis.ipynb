# Paso 5. Prueba de hipótesis

```python
# Revisión información contenida en el data_3
print(data_3.info())
data_3.head(10)
```

**Hipótesis nula**: La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos es igual a la duración promedio de los viajes den días no lluviosos.  

**Hipótesis alternativa**: La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos es diferente a la duración promedio de los viajes den días no lluviosos.  

```python
rainy_days = data_3[data_3['weather_conditions'] == 'Bad']
rainy_days
```

```python
no_rainy_days = data_3[data_3['weather_conditions'] == 'Good']
no_rainy_days
```

```python
rainy = rainy_days['duration_seconds']
no_rainy = no_rainy_days['duration_seconds']

# el valor de nivel de significancia es de 5%
alpha = 0.05

# prueba levene
levene_test = st.levene(rainy, no_rainy)
print('Resultado de la prueba de Levene (p-valor):', levene_test.pvalue)

# Verificación del resultado de la prueba de Levene
if levene_test.pvalue < alpha:
    print("Las varianzas son significativamente diferentes.")
    equal_var = False
else:
    print("Las varianzas no son significativamente diferentes.")
    equal_var = True


# Hipótesis
result = st.ttest_ind(
    no_rainy,
    rainy,
    equal_var = equal_var
)

print('valor p: ', result.pvalue)

# verificación del resultado
if result.pvalue < alpha:
    print("Rechazo la hipótesis nula. La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos es diferente.\n")
else:
    print("No rechazo la hipótesis nula. No hay evidencia suficiente para decir que La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos es igual.\n")
```
